```python
import pandas as pd
import numpy as np # ici j'importe mes bibliotheques
import matplotlib.pyplot as plt

df = pd.read_csv("FAOSTAT_data_en_11-18-2025.csv") # on a lu notre fichier

# utilisation de groupby pour calcule des moyenne
valeurs_moy_gby = df.groupby('Year')['Value'].mean()


# Faire notre propre groupby
# utlistion de la boucle (for) et de indexation (np.where) pour calcule des moyennes
annees = df['Year'].values  # Extraction des années en tableau numpy
annees_u = np.unique(annees)  # Obtention des années uniques
valeurs = df['Value'].values  # Extraction des valeurs en tableau numpy

resultats = []  # Liste pour stocker les moyennes calculées
for i in annees_u:
    index = np.where(annees==i)  # Trouve les indices où l'année est égale à i
    moyennes = np.nanmean(valeurs[index])  # Calcule la moyenne (ignorant les NaN)
    resultats.append(moyennes)  # Ajoute la moyenne à la liste des résultats
    print(i, moyennes)  # Affiche l'année et sa moyenne
```

```python
# Pour le code de visualisation suivant, il serait plus précis de calculer les points min/max comme ceci:
# Trouver l'index de la valeur minimale et maximale
# idx_min = np.argmin(resultats)
# idx_max = np.argmax(resultats)

# # Obtenir les années et valeurs correspondantes
# yr_min = annees_u[idx_min]
# val_min = resultats[idx_min]
# yr_max = annees_u[idx_max]
# val_max = resultats[idx_max]

# Define the missing variables first
yr_min = min(annees_u)  # Year with minimum value
val_min = min(resultats)  # Minimum value
yr_max = max(annees_u)  # Year with maximum value
val_max = max(resultats)  # Maximum value

# Now trace the plot with all variables defined
plt.plot(annees_u, resultats, 'o-k', linewidth=3)
plt.plot(yr_min, val_min, 'or', linewidth=5)
plt.plot(yr_max, val_max, 'ob', linewidth=5)
plt.xlabel('annees')
plt.ylabel('Moyennes annuelles')
```
